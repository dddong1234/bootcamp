환자의 기본정보를 변수에 저장하고, 다양한 연산자를 활용하여 의료 정보를 계산하는 프로그램을 작성해보세요.

```

출생연도: 1990년
성인 여부: True
```

### 힌트

- 현재 연도에서 출생연도를 빼면 나이가 나옵니다
- 논리 연산자로 성인 여부를 확인하세요

````python
# @title 정답

from datetime import datetime

# 1. 환자 기본정보 저장
patient_name = "박건강"      # 환자 이름
birth_year = 1990           # 출생연도
height_cm = 170             # 키 (cm)
weight = 65                 # 몸무게 (kg)
gender = "남성"             # 성별

# 2. BMI 계산
height_m = height_cm / 100              # 키를 미터로 변환
bmi = weight / (height_m ** 2)          # BMI 계산

# 3. 현재 나이 계산
current_year = datetime.now().year      # 현재 연도 자동 계산
age = current_year - birth_year         # 현재 나이 계산

# 4. 성인 여부 판별
is_adult = age >= 18                    # 성인 여부 (18세 이상)

# 5. 결과 출력
print("=== 환자 정보 ===")
print(f"이름: {patient_name} ({gender}, {age}세)")
print(f"키: {height_cm}cm, 몸무게: {weight}kg")
print(f"BMI: {bmi:.1f}")
print(f"출생연도: {birth_year}년")
print(f"성인 여부: {is_adult}")
````
