# 실습과제: 병원 환자 데이터베이스 관리

병원에서 제공하는 기본 환자 데이터베이스를 활용하여 환자 정보를 조회, 수정, 추가, 삭제하는 프로그램을 작성해보세요.

### 요구사항

1. **환자 정보 조회**
   - 첫 번째 환자의 모든 정보 출력 (get() 메서드 활용)
   - 딕셔너리 키별로 안전한 데이터 접근

2. **환자 정보 수정**
   - 두 번째 환자의 질병 정보 수정 (update() 메서드 활용)
   - 수정 전후 상태 비교 출력

3. **새 환자 등록**
   - 새로운 환자 데이터를 딕셔너리로 생성 후 데이터베이스에 추가
   - 추가된 환자 정보를 데이터베이스에서 다시 조회하여 출력

4. **환자 퇴원 처리**
   - 마지막 환자를 데이터베이스에서 제거
   - 퇴원한 환자의 정보 출력

5. **통계 정보**
   - 환자들의 나이 통계 (최고령, 최연소, 평균)
   - 내장 함수 활용 (max, min, sum)

6. **출력 형식**
```
=== 첫 번째 환자 정보 ===
환자번호: P001
이름: 김환자
나이: 65세
진료과: 내과
질병: 고혈압

=== 환자 질병 수정 전 ===
환자번호: P002
이름: 이환자
기존 질병: 맹장염

=== 환자 질병 수정 후 ===
환자번호: P002
이름: 이환자
수정된 질병: 급성맹장염

=== 새 환자 등록 ===
환자번호: P004
이름: 홍환자
나이: 40세
진료과: 피부과
질병: 아토피
현재 총 환자 수: 4명

=== 환자 퇴원 처리 ===
퇴원 환자번호: P004
퇴원 환자명: 홍환자
퇴원 후 총 환자 수: 3명

=== 환자 나이 통계 ===
최고령 환자 나이: 65세
최연소 환자 나이: 32세
평균 나이: 47.3세

==================================================
최종 병원 환자 데이터베이스 현황
==================================================
총 입원 환자: 3명
```

### 힌트

- `get()` 메서드를 사용해 안전하게 딕셔너리 값에 접근하세요
- `update()` 메서드를 사용해 딕셔너리 정보를 수정하세요
- `append()` 메서드로 새 환자를 리스트에 추가하세요
- `pop()` 메서드로 마지막 환자를 제거하세요
- `len()` 함수로 총 환자 수를 계산하세요
- 새 환자 정보 출력 시 `patients_db[-1]`로 접근하세요


## 💻 구현해보세요

````python
# 기본 데이터 - 병원 환자 데이터베이스
patients_db = [
    {"patient_id": "P001", "name": "김환자", "age": 65, "department": "내과", "disease": "고혈압"},
    {"patient_id": "P002", "name": "이환자", "age": 45, "department": "외과", "disease": "맹장염"},
    {"patient_id": "P003", "name": "박환자", "age": 32, "department": "소아과", "disease": "감기"}
]
````

````python
# 여기에 코드를 작성하세요
# 1. 첫 번째 환자 정보 조회
print("=== 첫 번째 환자 정보 ===")
print(f"환자번호: {patients_db[0].get("patient_id")}")
print(f"이름: {patients_db[0].get("name")}")
print(f"나이: {patients_db[0].get("age")}세")
print(f"진료과: {patients_db[0].get("department")}")
print(f"질병: {patients_db[0].get("disease")}\n")

# 2. 두 번째 환자 질병 수정
print("=== 환자 질병 수정 전 ===")
print(f"환자번호: {patients_db[1].get("patient_id")}")
print(f"이름: {patients_db[1].get("name")}")
print(f"기존 질병: {patients_db[1].get("disease")}\n")

patients_db[1].update({"disease":"급성맹장염"})

print("=== 환자 질병 수정 후 ===")
print(f"환자번호: {patients_db[1].get("patient_id")}")
print(f"이름: {patients_db[1].get("name")}")
print(f"수정된 질병: {patients_db[1].get("disease")}\n")

# 3. 새 환자 등록 및 데이터베이스에서 조회
patients_db.append({
    "patient_id": "P004",
    "name": "홍환자",
    "age": 40,
    "department": "피부과",
    "disease": "아토피"
})
print("=== 새 환자 등록 ===")
print(f"환자번호: {patients_db[-1].get("patient_id")}")
print(f"이름: {patients_db[-1].get("name")}")
print(f"나이: {patients_db[-1].get("age")}세")
print(f"진료과: {patients_db[-1].get("department")}")
print(f"질병: {patients_db[-1].get("disease")}")
print(f"현재 총 환자 수: {len(patients_db)}명\n")


# 4. 환자 퇴원 처리
discharged_patient = patients_db.pop()
print("=== 환자 퇴원 처리 ===")
print(f"환자번호: {discharged_patient.get("patient_id")}")
print(f"이름: {discharged_patient.get("name")}")
print(f"퇴원 후 총 환자 수: {len(patients_db)}명\n")

# 5. 나이 통계 계산
ages = []
for i in patients_db:
  ages.append(i.get("age"))

max_age = max(ages)
min_age = min(ages)
avg_age = sum(ages)/len(ages)

print(f"최고령 환자 나이: {max_age}세")
print(f"최연소 환자 나이: {min_age}세")
print(f"평균 나이: {avg_age:.1f}세\n")

# 6. 최종 현황 출력

print("==================================================")
print("최종 병원 환자 데이터베이스 현황")
print("==================================================\n")

print(f"총 입원 환자: {len(patients_db)}")
````

````python

# @title 정답

# 기본 데이터 - 병원 환자 데이터베이스
patients_db = [
    {"patient_id": "P001", "name": "김환자", "age": 65, "department": "내과", "disease": "고혈압"},
    {"patient_id": "P002", "name": "이환자", "age": 45, "department": "외과", "disease": "맹장염"},
    {"patient_id": "P003", "name": "박환자", "age": 32, "department": "소아과", "disease": "감기"}
]

# 1. 특정 환자 정보 조회 (첫 번째 환자) - get() 메서드 활용
first_patient = patients_db[0]
print("=== 첫 번째 환자 정보 ===")
print(f"환자번호: {first_patient.get('patient_id', '정보없음')}")
print(f"이름: {first_patient.get('name', '정보없음')}")
print(f"나이: {first_patient.get('age', 0)}세")
print(f"진료과: {first_patient.get('department', '정보없음')}")
print(f"질병: {first_patient.get('disease', '정보없음')}")

# 2. 특정 환자 질병 수정 (두 번째 환자) - update() 메서드 활용
second_patient = patients_db[1]
print(f"\n=== 환자 질병 수정 전 ===")
print(f"환자번호: {second_patient.get('patient_id', '정보없음')}")
print(f"이름: {second_patient.get('name', '정보없음')}")
print(f"기존 질병: {second_patient.get('disease', '정보없음')}")

second_patient.update({"disease": "급성맹장염"})  # update() 메서드로 질병 수정

print(f"\n=== 환자 질병 수정 후 ===")
print(f"환자번호: {second_patient.get('patient_id', '정보없음')}")
print(f"이름: {second_patient.get('name', '정보없음')}")
print(f"수정된 질병: {second_patient.get('disease', '정보없음')}")

# 3. 새로운 환자 추가
new_patient = {"patient_id": "P004", "name": "홍환자", "age": 40, "department": "피부과", "disease": "아토피"}
patients_db.append(new_patient)

# 새로 추가된 환자 정보를 데이터베이스에서 조회 (patients_db[-1] 활용)
latest_patient = patients_db[-1]
print(f"\n=== 새 환자 등록 ===")
print(f"환자번호: {latest_patient.get('patient_id', '정보없음')}")
print(f"이름: {latest_patient.get('name', '정보없음')}")
print(f"나이: {latest_patient.get('age', 0)}세")
print(f"진료과: {latest_patient.get('department', '정보없음')}")
print(f"질병: {latest_patient.get('disease', '정보없음')}")
print(f"현재 총 환자 수: {len(patients_db)}명")

# 4. 환자 퇴원 처리 (마지막 환자 제거)
discharged_patient = patients_db.pop()
print(f"\n=== 환자 퇴원 처리 ===")
print(f"퇴원 환자번호: {discharged_patient.get('patient_id', '정보없음')}")
print(f"퇴원 환자명: {discharged_patient.get('name', '정보없음')}")
print(f"퇴원 후 총 환자 수: {len(patients_db)}명")

# 5. 통계 함수 활용 (반복문/조건문 없이)
ages = [patients_db[0]["age"], patients_db[1]["age"], patients_db[2]["age"]]
print(f"\n=== 환자 나이 통계 ===")
print(f"최고령 환자 나이: {max(ages)}세")
print(f"최연소 환자 나이: {min(ages)}세")
print(f"평균 나이: {sum(ages) / len(ages):.1f}세")

# 6. 최종 데이터베이스 현황
print(f"\n{'='*50}")
print("최종 병원 환자 데이터베이스 현황")
print(f"{'='*50}")
print(f"총 입원 환자: {len(patients_db)}명")
````
