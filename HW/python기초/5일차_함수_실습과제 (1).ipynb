# 실습과제: 환자 예약 관리 시스템

병원에서 환자들의 진료 예약을 관리하는 시스템을 작성해보세요. 예약 시간 중복을 방지하고 운영시간을 체크하는 실무적인 예약 시스템을 구현합니다.

## 요구사항

1. **예약 가능 여부 확인 함수**
   - 운영시간 내 예약인지 확인 (9-18시)
   - 점심시간 제외 (12-13시)
   - 이미 예약된 시간인지 확인 (중복 방지)
   - 결과를 (가능여부, 상태키워드) 튜플로 반환

2. **예약 추가 함수**
   - 예약 가능하면 예약 리스트에 딕셔너리 형태로 추가
   - 예약 불가능하면 구체적인 실패 이유 반환
   - 결과 메시지를 문자열로 반환

3. **예약 현황 출력 함수**
   - 현재 예약된 환자들을 시간순으로 출력
   - 예약이 없으면 "예약된 환자가 없습니다" 출력

## 출력 형식
```
=== 예약 처리 결과 ===
김환자 - 10시: 예약 완료
이환자 - 10시: 이미 예약된 시간입니다
박환자 - 12시: 점심시간 (예약 불가)
정환자 - 20시: 운영시간 외 (예약 불가)
최환자 - 14시: 예약 완료

=== 현재 예약 현황 ===
10시: 김환자
14시: 최환자
```

## 📋 데이터 구조 설명

### **appointments 리스트**
딕셔너리를 요소로 하는 리스트입니다.
```python
appointments = [
    {"time": 10, "patient": "김환자"},
    {"time": 14, "patient": "최환자"}
]
```
- 각 딕셔너리는 하나의 예약을 나타냄
- `"time"`: 예약 시간 (정수)
- `"patient"`: 환자 이름 (문자열)

## 힌트

- `clinic_info` 딕셔너리에는 운영시간과 점심시간 정보가 튜플로 저장되어 있습니다
- 상태 키워드 상수(AVAILABLE, LUNCH_TIME 등)로 예약 상태를 관리하세요
- 튜플 언패킹으로 함수 반환값을 받으세요: `is_available, status = 함수()`
- 딕셔너리 접근으로 예약 정보를 관리하세요: `appointment["time"]`
- 리스트의 append() 메서드로 새 예약을 추가하세요


## 💻 구현해보세요


````python
# 기본 데이터
clinic_info = {
    "operating_hours": (9, 18),    # 클리닉 운영시간: 9시-18시
    "lunch_break": (12, 13),       # 점심시간: 12시-13시
}

# 예약 현황 리스트 (전역 변수)
appointments = []  # 딕셔너리 리스트: [{"time": 10, "patient": 김환자}, ...]

# 상태 키워드
AVAILABLE = "AVAILABLE"           # 예약 가능
LUNCH_TIME = "LUNCH_TIME"         # 점심시간
OUT_OF_HOURS = "OUT_OF_HOURS"     # 운영시간 외
ALREADY_BOOKED = "ALREADY_BOOKED" # 이미 예약됨
````

````python

START_TIME , END_TIME = clinic_info["operating_hours"]
LUNCH_START, LUNCH_END = clinic_info["lunch_break"]

def get_booked_time():
    return {appointment.get("time") for appointment in appointments}

def check_appointment_availability(time):
    booked_time = get_booked_time()
    status = AVAILABLE
    if not START_TIME <= time < END_TIME:
        status = OUT_OF_HOURS
        return False, status
    if LUNCH_START <= time <= LUNCH_END:
        status = LUNCH_TIME
        return False, status
    if time in booked_time:
        status = ALREADY_BOOKED
        return False, status
    return True, status


def add_appointment(patient_name, time):
    is_available, status = check_appointment_availability(time)
    appointment_str = ""
    match status:
        case "AVAILABLE" :
            appointment_str = "예약 완료"
        case "LUNCH_TIME" :
            appointment_str = "점심시간 (예약 불가)"
        case "OUT_OF_HOURS" :
            appointment_str = "운영시간 외 (예약 불가)"
        case "ALREADY_BOOKED" :
            appointment_str = "이미 예약된 시간입니다"

    if is_available:
        appointments.append({"patient_name": patient_name, "time": time})
    return f"{patient_name} - {time}시: {appointment_str}"

def show_appointments():
    if not appointments:
        print("예약된 환자가 없습니다.")
        return
    for appointment in sorted(appointments, key=lambda x : x["time"]):
        print(f"{appointment['time']}시: {appointment['patient_name']}")

````

````python
# 테스트 데이터
test_requests = [
    ("김환자", 10),    # 정상 예약
    ("이환자", 10),    # 중복 시간
    ("박환자", 12),    # 점심시간
    ("정환자", 20),    # 운영시간 외
    ("최환자", 14)     # 정상 예약
]
````

````python
# 테스트 실행
print("=== 예약 처리 결과 ===")
for patient_name, hour in test_requests:
    result = add_appointment(patient_name, hour)
    print(result)

print("\n=== 현재 예약 현황 ===")
show_appointments()
````

````python
# @title 정답

# 기본 데이터
clinic_info = {
    "operating_hours": (9, 18),    # 클리닉 운영시간: 9시-18시
    "lunch_break": (12, 13),       # 점심시간: 12시-13시
}

# 예약 현황 리스트 (전역 변수)
appointments = []  # 딕셔너리 리스트: [{"time": 10, "patient": 김환자}, ...]

# 상태 키워드
AVAILABLE = "AVAILABLE"           # 예약 가능
LUNCH_TIME = "LUNCH_TIME"         # 점심시간
OUT_OF_HOURS = "OUT_OF_HOURS"     # 운영시간 외
ALREADY_BOOKED = "ALREADY_BOOKED" # 이미 예약됨

def check_appointment_availability(hour):
    """
    예약 가능 여부를 종합적으로 확인합니다.

    Args:
        hour (int): 예약 시간 (0-23)

    Returns:
        tuple: (가능여부(bool), 키워드(str))
    """
    # 1단계: 운영시간 확인
    start, end = clinic_info["operating_hours"]
    if not (start <= hour < end):
        return False, OUT_OF_HOURS

    # 2단계: 점심시간 확인
    lunch_start, lunch_end = clinic_info["lunch_break"]
    if lunch_start <= hour < lunch_end:
        return False, LUNCH_TIME

    # 3단계: 예약 중복 확인
    for appointment in appointments:
        if appointment["time"] == hour:
            return False, ALREADY_BOOKED

    # 모든 검증 통과
    return True, AVAILABLE

def add_appointment(patient_name, hour):
    """
    새로운 예약을 추가합니다.

    Args:
        patient_name (str): 환자 이름
        hour (int): 예약 시간 (0-23)

    Returns:
        str: 예약 결과 메시지
    """
    is_available, status = check_appointment_availability(hour)

    if is_available:
        appointments.append({"time": hour, "patient": patient_name})
        return f"{patient_name} - {hour}시: 예약 완료"
    else:
        if status == LUNCH_TIME:
            reason = "점심시간 (예약 불가)"
        elif status == OUT_OF_HOURS:
            reason = "운영시간 외 (예약 불가)"
        elif status == ALREADY_BOOKED:
            reason = "이미 예약된 시간입니다"

        return f"{patient_name} - {hour}시: {reason}"

def show_appointments():
    """
    현재 예약 현황을 보기 좋게 출력합니다.

    Returns:
        None: 예약 현황을 출력만 함
    """
    if not appointments:
        print("예약된 환자가 없습니다.")
        return

    for appointment in appointments:
        time = appointment["time"]
        patient = appointment["patient"]
        print(f"{time}시: {patient}")

# 테스트 데이터
test_requests = [
    ("김환자", 10),    # 정상 예약
    ("이환자", 10),    # 중복 시간
    ("박환자", 12),    # 점심시간
    ("정환자", 20),    # 운영시간 외
    ("최환자", 14)     # 정상 예약
]

# 테스트 실행
print("=== 예약 처리 결과 ===")
for patient_name, hour in test_requests:
    result = add_appointment(patient_name, hour)
    print(result)

print("\n=== 현재 예약 현황 ===")
show_appointments()
````
