# 실습과제: 의료진 근무 관리 시스템

병원에서 의료진들의 근무 현황을 관리하는 시스템을 객체지향으로 구현해보세요. 의료진별 근무 상태를 체크하고 환자 배정을 관리하는 실무적인 시스템을 만들어봅시다.


## 요구사항


### **상속 구조**
- `Nurse` 클래스: MedicalStaff 상속 + 담당환자 관리
- `Doctor` 클래스: MedicalStaff 상속 + 전문분야 + 진료기능


### **1. MedicalStaff 클래스 (부모 클래스)**
의료진의 기본 정보와 공통 기능을 담당하는 부모 클래스입니다.

**속성**
- `name`: 의료진 이름 (문자열)
- `department`: 근무 부서 (문자열: "내과", "외과", "소아과", "응급실" 등)
- `status`: 근무 상태 (문자열: "근무중"으로 초기화)

**메서드**
- `change_status(new_status)`: 근무 상태를 변경
- `get_info()`: 의료진 기본 정보를 문자열로 반환

```python
# 예시 구조
class MedicalStaff:
    def __init__(self, name, department):
        # 속성 초기화
        
    def change_status(self, new_status):
        # 상태 변경 로직
        
    def get_info(self):
        # 정보 반환 로직
        return "김의사(내과) - 근무중"
```

### **2. Nurse 클래스 (MedicalStaff 상속)**
의료진 클래스를 상속받아 간호사만의 고유 기능을 추가한 클래스입니다.

**추가 속성**
- `patients`: 담당 환자 리스트 (빈 리스트로 초기화)

**메서드**
- `assign_patient(patient_name)`: 환자를 담당 리스트에 배정
  - 근무 상태가 "근무중"이 아니면 배정 불가
  - 이미 담당 중인 환자면 중복 배정 방지
- `remove_patient(patient_name)`: 담당 환자를 리스트에서 제거

```python
# 예시 구조
class Nurse(MedicalStaff):
    def __init__(self, name, department):
        super().__init__(name, department)  # 부모 클래스 초기화
        # 간호사 고유 속성 추가
        
    def assign_patient(self, patient_name):
        # 근무 상태 확인 → 중복 확인 → 배정 처리
        return "김간호사에게 환자A 배정 완료"
```

### **3. Doctor 클래스 (MedicalStaff 상속)**
의료진 클래스를 상속받아 의사만의 전문 기능을 구현한 클래스입니다.

**추가 속성**
- `specialty`: 전문 분야 (문자열: "심장내과", "정형외과", "신경외과" 등)

**메서드**
- `treat_patient(patient_name, diagnosis)`: 환자 진료 기능
  - 근무 상태가 "근무중"이 아니면 진료 불가
  - 진료 완료 시 결과 메시지 반환

```python
# 예시 구조  
class Doctor(MedicalStaff):
    def __init__(self, name, department, specialty):
        super().__init__(name, department)  # 부모 클래스 초기화
        # 의사 고유 속성 추가
        
    def treat_patient(self, patient_name, diagnosis):
        # 근무 상태 확인 → 진료 처리
        return "이의사가 환자D을(를) 무릎 관절염로 진료했습니다"
```

## 출력 형식
```
=== 환자 배정 결과 ===
김간호사에게 환자A 배정 완료
김간호사에게 환자B 배정 완료
이의사는 현재 휴무중이므로 환자 배정 불가

=== 의료진 현황 ===
김간호사(내과) - 근무중
이의사(외과) - 휴무중

=== 진료 결과 ===
이의사가 환자D을(를) 무릎 관절염로 진료했습니다.
```

## 힌트

- 클래스명은 **PascalCase**를 사용하세요 (MedicalStaff, Nurse, Doctor)
- 생성자에서 **super()** 를사용해 부모 클래스 초기화를 호출하세요
- 리스트의 **append()와 remove()** 메서드를 사용해 환자 관리하세요
- **조건문**으로 근무 상태를 확인한 후 기능을 수행하세요
- **f-string**을 사용해 메시지를 보기 좋게 만드세요

````python
class MedicalStaff:
    def __init__(self,name:str,department:str):
        self.name = name
        self.department = department
        self.status = "근무중"

    def change_status(self,new_status):
        self.status = new_status

    def get_info(self):
        return f"{self.name}({self.department}) - {self.status}"

class Nurse(MedicalStaff):
    def __init__(self,name:str,department:str):
        super().__init__(name,department)
        self.patients = []

    def assign_patient(self,patient_name:str):
        if self.status != "근무중":
            return f"{self.name} {self.status}, 환자 배정 불가"
        if patient_name in self.patients:
            return f"{patient_name}은(는) {self.name}가 이미 담당 중인 환자입니다."
        self.patients.append(patient_name)
        return f"{self.name}에게 {patient_name} 배정 완료"

class Doctor(MedicalStaff):
    def __init__(self,name:str,department:str,specialty:str):
        super().__init__(name,department)
        self.specialty = specialty

    def treat_patient(self,patient_name:str,diagnosis:str):
        if self.status != "근무중":
            return f"{self.name} {self.status}. 진료 불가"
        return f"{self.name}가 {patient_name}을(를) {diagnosis}로 진료했습니다"
````

````python
# 테스트 데이터
# 의료진 생성
nurse = Nurse("김간호사", "내과")
doctor = Doctor("이의사", "외과", "정형외과")

# 환자 배정
print("=== 환자 배정 결과 ===")
print(nurse.assign_patient("환자A"))
print(nurse.assign_patient("환자B"))

# 의사 상태 변경
doctor.change_status("휴무중")
print("이의사는 현재 휴무중이므로 환자 배정 불가")

# 현재 상황 출력
print("\n=== 의료진 현황 ===")
print(nurse.get_info())
print(doctor.get_info())

# 의사 근무 복귀 후 진료
doctor.change_status("근무중")
print("\n=== 진료 결과 ===")
print(doctor.treat_patient("환자D", "무릎 관절염"))
````

````python
# @title 정답

class MedicalStaff:
    """의료진 기본 정보와 공통 기능을 관리하는 부모 클래스"""

    def __init__(self, name, department):
        """
        의료진 기본 정보를 초기화합니다.

        Args:
            name (str): 의료진 이름
            department (str): 근무 부서
        """
        self.name = name                # 의료진 이름
        self.department = department    # 근무 부서
        self.status = "근무중"          # 근무 상태 (기본값: 근무중)

    def change_status(self, new_status):
        """
        근무 상태를 변경합니다.

        Args:
            new_status (str): 새로운 근무 상태

        Returns:
            None
        """
        self.status = new_status

    def get_info(self):
        """
        의료진 기본 정보를 반환합니다.

        Args:
            None

        Returns:
            str: "이름(부서) - 상태" 형식의 정보
        """
        return f"{self.name}({self.department}) - {self.status}"


class Nurse(MedicalStaff):
    """간호사 정보와 환자 담당 관리 기능을 제공하는 클래스"""

    def __init__(self, name, department):
        """
        간호사 정보를 초기화합니다.

        Args:
            name (str): 간호사 이름
            department (str): 근무 부서
        """
        super().__init__(name, department)  # 부모 클래스 초기화
        self.patients = []                  # 담당 환자 리스트 (빈 리스트로 초기화)

    def assign_patient(self, patient_name):
        """
        환자를 담당 리스트에 배정합니다.

        Args:
            patient_name (str): 환자 이름

        Returns:
            str: 배정 결과 메시지
        """
        # 근무 상태 확인
        if self.status != "근무중":
            return f"{self.name}은(는) 현재 {self.status}이므로 환자 배정 불가"

        # 중복 배정 확인 및 처리
        if patient_name not in self.patients:
            self.patients.append(patient_name)
            return f"{self.name}에게 {patient_name} 배정 완료"
        else:
            return f"{patient_name}은(는) 이미 {self.name}가 담당 중입니다"

    def remove_patient(self, patient_name):
        """
        담당 환자를 리스트에서 제거합니다.

        Args:
            patient_name (str): 환자 이름

        Returns:
            str: 제거 결과 메시지
        """
        # 담당 환자 여부 확인 및 제거 처리
        if patient_name in self.patients:
            self.patients.remove(patient_name)
            return f"{self.name}의 담당에서 {patient_name} 제거 완료"
        else:
            return f"{patient_name}은(는) {self.name}의 담당 환자가 아닙니다"


class Doctor(MedicalStaff):
    """의사 정보와 전문 진료 기능을 제공하는 클래스"""

    def __init__(self, name, department, specialty):
        """
        의사 정보를 초기화합니다.

        Args:
            name (str): 의사 이름
            department (str): 근무 부서
            specialty (str): 전문 분야
        """
        super().__init__(name, department)  # 부모 클래스 초기화
        self.specialty = specialty          # 전문 분야 (심장내과, 정형외과 등)

    def treat_patient(self, patient_name, diagnosis):
        """
        환자를 진료하고 결과를 반환합니다.

        Args:
            patient_name (str): 환자 이름
            diagnosis (str): 진단명

        Returns:
            str: 진료 결과 메시지
        """
        # 근무 상태 확인
        if self.status != "근무중":
            return f"{self.name} 의사는 현재 {self.status}이므로 진료 불가"

        # 진료 처리
        return f"{self.name}가 {patient_name}을(를) {diagnosis}로 진료했습니다"


# 테스트 데이터
# 의료진 생성
nurse = Nurse("김간호사", "내과")
doctor = Doctor("이의사", "외과", "정형외과")

# 환자 배정
print("=== 환자 배정 결과 ===")
print(nurse.assign_patient("환자A"))
print(nurse.assign_patient("환자B"))

# 의사 상태 변경
doctor.change_status("휴무중")
print("이의사는 현재 휴무중이므로 환자 배정 불가")

# 현재 상황 출력
print("\n=== 의료진 현황 ===")
print(nurse.get_info())
print(doctor.get_info())

# 의사 근무 복귀 후 진료
doctor.change_status("근무중")
print("\n=== 진료 결과 ===")
print(doctor.treat_patient("환자D", "무릎 관절염"))
````
