# 실습과제: 환자 진료기록 파일 관리 시스템

병원에서 환자들의 진료기록을 안전하게 파일로 저장하고 읽어오는 시스템을 작성해보세요. 파일 입출력과 예외처리를 활용하여 의료 현장에서 필요한 환자 기록 관리 기능을 구현합니다.

## 요구사항

1. **파일 저장 기능**
   - 환자 ID, 이름, 진단명을 받아서 개별 파일로 저장
   - 환자 ID와 이름이 비어있으면 ValueError 발생시키기
   - 파일명 형식: `{patient_id}.txt`
   - 성공 시: "✅ {이름} 환자 저장 완료" 출력

2. **파일 읽기 기능**
   - 환자 ID로 해당 파일 내용 반환
   - 파일이 없으면 FileNotFoundError 처리
   - 성공 시: "✅ {patient_id} 환자 조회 완료" 출력

3. **출력 형식**
```
=== 환자 등록 테스트 ===
✅ 김환자 환자 저장 완료
✅ 이환자 환자 저장 완료
❌ 오류: 환자 ID와 이름은 필수입니다
❌ 오류: 환자 ID와 이름은 필수입니다
✅ 최환자 환자 저장 완료

=== 환자 조회 테스트 ===
✅ P001 환자 조회 완료
환자명: 김환자
진단: 감기

✅ P002 환자 조회 완료
환자명: 이환자
진단: 두통

❌ 환자 P999를 찾을 수 없습니다
```

## 힌트

- `try-except` 블록을 사용해 예외를 안전하게 처리하세요
- `with open()` 구문으로 파일을 자동으로 닫도록 하세요
- `ValueError`와 `FileNotFoundError`를 구분해서 처리하세요
- `for` 반복문으로 테스트 데이터를 순회하세요
- f-string을 사용해 파일명과 메시지를 깔끔하게 작성하세요
- `if not` 조건문으로 빈 값을 검증하세요


## 💻 구현해보세요

````python
def save_patient(patient_id, name, diagnosis):
    try:
        if not patient_id or not name:
            raise ValueError("환자 ID와 이름은 필수입니다")

        with open(f"{patient_id}.txt", "w", encoding = 'utf-8') as patients_file:
            patients_file.write(f"환자명: {name}\n진단: {diagnosis}\n")
            print(f"✅ {name} 환자 저장 완료")
    except ValueError as e:
        print(f"❌ 오류: {e}")
    except OSError as e:
        print(f"❌ 파일 처리 오류: {e}")




def load_patient(patient_id):
    try:
        with open(f"{patient_id}.txt", "r", encoding = 'utf-8') as patients_file:
            print(f"✅ {patient_id} 환자 조회 완료")
            return patients_file.read()
    except FileNotFoundError:
        print(f"❌ 환자 {patient_id}를 찾을 수 없습니다.")
````

````python
# 테스트 데이터
test_patients = [
    ("P001", "김환자", "감기"),
    ("P002", "이환자", "두통"),
    ("", "박환자", "복통"),      # 오류 케이스: 빈 ID
    ("P003", "", "발열"),        # 오류 케이스: 빈 이름
    ("P004", "최환자", "고혈압")
]

test_lookups = ["P001", "P002", "P999"]  # P999는 없는 환자
````

````python
# 1. 환자 등록 테스트
print("=== 환자 등록 테스트 ===")
for patient_id, name, diagnosis in test_patients:
    save_patient(patient_id, name, diagnosis)

# 2. 환자 조회 테스트
print("\n=== 환자 조회 테스트 ===")
for patient_id in test_lookups:
    content = load_patient(patient_id)
    if content:
        print(content)
````

````python
# @title 정답

def save_patient(patient_id, name, diagnosis):
    """환자 정보를 파일에 저장합니다"""
    try:
        # 입력값 검증
        if not patient_id or not name:
            raise ValueError("환자 ID와 이름은 필수입니다") # ValueError 에러 발생시키기

        # 파일에 저장
        filename = f"{patient_id}.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(f"환자명: {name}\n")
            file.write(f"진단: {diagnosis}\n")

        print(f"✅ {name} 환자 저장 완료")
        return True

    except ValueError as e:
        print(f"❌ 오류: {e}")
        return False
    except Exception as e:
        print(f"❌ 저장 실패: {e}")
        return False

def load_patient(patient_id):
    """환자 정보를 파일에서 읽습니다"""
    try:
        filename = f"{patient_id}.txt"
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()

        print(f"✅ {patient_id} 환자 조회 완료")
        return content

    except FileNotFoundError:
        print(f"❌ 환자 {patient_id}를 찾을 수 없습니다")
        return None
    except Exception as e:
        print(f"❌ 조회 실패: {e}")
        return None

# 테스트 데이터
test_patients = [
    ("P001", "김환자", "감기"),
    ("P002", "이환자", "두통"),
    ("", "박환자", "복통"),      # 오류 케이스: 빈 ID
    ("P003", "", "발열"),        # 오류 케이스: 빈 이름
    ("P004", "최환자", "고혈압")
]

test_lookups = ["P001", "P002", "P999"]  # P999는 없는 환자

# 1. 환자 등록 테스트
print("=== 환자 등록 테스트 ===")
for patient_id, name, diagnosis in test_patients:
    save_patient(patient_id, name, diagnosis)

# 2. 환자 조회 테스트
print("\n=== 환자 조회 테스트 ===")
for patient_id in test_lookups:
    content = load_patient(patient_id)
    if content:
        print(content)
````
