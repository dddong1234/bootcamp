# 실습과제: 환자 데이터 분석 연습

병원에서 수집한 환자 데이터를 Pandas를 사용하여 분석해보세요.   
DataFrame 생성부터 조건부 데이터 추출, 기본 통계까지 실무에서 자주 사용하는 핵심 기능들을 연습합니다.

## 요구사항

1. **환자 정보 DataFrame 생성**
   - 주어진 딕셔너리 데이터로 DataFrame 생성
   - 환자ID를 인덱스로 설정
   - 생성된 DataFrame의 기본 정보 확인 (환자 수, 컬럼 수)

2. **특정 환자 정보 조회 (loc, iloc 활용)**
   - 'P003' 환자의 모든 정보 조회 (loc 사용)
   - 첫 번째와 세 번째 환자 정보 조회 (iloc 사용)
   - '나이'와 '혈압' 컬럼만 선택하여 조회

3. **조건부 데이터 필터링**
   - 고혈압 환자 (혈압 ≥ 140) 찾기
   - 65세 이상 고령 환자 찾기
   - 내과 환자 중 혈당이 높은 환자 (혈당 ≥ 120) 찾기

4. **기본 통계 분석**
   - 진료과별 환자 수 계산 (value_counts)
   - 성별 분포 확인 (value_counts)

5. **위험도 등급 추가**
   - 고혈압이면 '고위험', 아니면 '정상'으로 새 컬럼 생성
   - 위험도별 환자 분포 확인

## 힌트

- DataFrame 생성: `pd.DataFrame(딕셔너리).set_index('컬럼명')`
- 특정 행 조회: `df.loc['인덱스']`, `df.iloc[숫자]`
- 조건 필터링: `df[df['컬럼'] >= 값]`
- 여러 조건: `df[(df['컬럼1'] == 값1) & (df['컬럼2'] >= 값2)]`
- 분포 확인: `df['컬럼'].value_counts()`
- 조건부 값 할당: `df.loc[조건, '새컬럼'] = '값'`

````python
import pandas as pd

# 기본 데이터
patient_data = {
    '환자ID': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006'],
    '이름': ['김철수', '이영희', '박민수', '최지은', '정도영', '한소희'],
    '나이': [45, 67, 32, 29, 58, 34],
    '성별': ['남', '여', '남', '여', '남', '여'],
    '혈압': [120, 145, 110, 115, 160, 125],
    '혈당': [95, 156, 88, 92, 102, 108],
    '진료과': ['내과', '내과', '소아과', '내과', '순환기내과', '소아과']
}

````

````python
# - DataFrame 생성: `pd.DataFrame(딕셔너리).set_index('컬럼명')`
# - 특정 행 조회: `df.loc['인덱스']`, `df.iloc[숫자]`
# - 조건 필터링: `df[df['컬럼'] >= 값]`
# - 여러 조건: `df[(df['컬럼1'] == 값1) & (df['컬럼2'] >= 값2)]`
# - 분포 확인: `df['컬럼'].value_counts()`
# - 조건부 값 할당: `df.loc[조건, '새컬럼'] = '값'`

# 1. DataFrame 생성 및 기본 정보
print("=== 1. DataFrame 생성 ===")
# DataFrame 생성 코드 작성
df = pd.DataFrame(patient_data)

# 환자ID를 인덱스로 설정
df.set_index('환자ID', inplace=True)

# 기본 정보 출력 (환자 수, 컬럼 수)
print(f"환자 수: {df.shape[0]}명")
print(f"컬럼 수: {df.shape[1]}개")
print(f"컬럼명: {df.columns.to_list()}")
print("\n생성된 DataFrame:")
print(df)

# 2. 특정 환자 정보 조회 (loc, iloc 활용)
print("\n=== 2. 특정 환자 조회 ===")
# P003 환자 정보 조회
patient_P003 = df.loc['P003']
print(f"P003 환자 정보:\n{patient_P003}")

# 첫 번째, 세 번째 환자 조회
patient_first_third = df.iloc[[0,2]]
print("\n첫 번째, 세 번째 환자 정보:")
print(patient_first_third)

# 나이, 혈압 컬럼만 선택하여 전체 조회
print("\n모든 환자의 나이와 혈압:")
print(df[['나이','혈압']])


# 3. 조건부 데이터 필터링
print("\n=== 3. 조건부 필터링 ===")
# 고혈압 환자 찾기 (혈압 >= 140)
htn_patients = df[df['혈압']>=140]
print(f"고혈압 환자 ({len(htn_patients)}명):")
print(htn_patients[['이름','혈압']])
# 고령 환자 찾기 (나이 >= 65)
old_patients = df[df['나이']>=65]
print(f"\n고령 환자 ({len(old_patients)}명):")
print(old_patients[['이름','나이']])
# 내과 환자 중 고혈당 환자 찾기 (진료과 == '내과' AND 혈당 >= 120)
dm_patients = df[(df['진료과'] == '내과') & (df['혈당'] >=120)]
print(f"\n내과 고혈당 환자 ({len(dm_patients)}명):")
print(dm_patients[['이름','진료과','혈당']])
# 4. 기본 통계 분석 (value_counts 중심)
print("\n=== 4. 기본 통계 ===")
# 진료과별 환자 수 계산 (value_counts)
dept_counts = df['진료과'].value_counts()
print("진료과별 환자 수:")
print(dept_counts)
# 성별 분포 확인 (value_counts)
gender_counts = df['성별'].value_counts()
print("\n성별 분포:")
print(gender_counts)
# 5. 위험도 등급 추가
print("\n=== 5. 위험도 등급 추가 ===")
# 고혈압이면 '고위험', 아니면 '정상'으로 새 컬럼 생성
htn_mask = df['혈압']>=140
df.loc[htn_mask, '위험도'] = '고위험'
df.loc[~htn_mask, '위험도'] = '정상'
#numpy의 where 메서드 활용가능할듯 합니다
#df['위험도'] = np.where(df['혈압']>=140,'고혈압','정상)

# 위험도별 환자 분포 확인 (value_counts)
risk_counts = df['위험도'].value_counts()
print("위험도별 환자 분포:")
print(risk_counts)

print("\n최종 DataFrame (위험도 컬럼 추가):")
print(df)
````

````python
# @title 정답

import pandas as pd

# 기본 데이터
patient_data = {
    '환자ID': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006'],
    '이름': ['김철수', '이영희', '박민수', '최지은', '정도영', '한소희'],
    '나이': [45, 67, 32, 29, 58, 34],
    '성별': ['남', '여', '남', '여', '남', '여'],
    '혈압': [120, 145, 110, 115, 160, 125],
    '혈당': [95, 156, 88, 92, 102, 108],
    '진료과': ['내과', '내과', '소아과', '내과', '순환기내과', '소아과']
}

# 1. DataFrame 생성 및 기본 정보
print("=== 1. DataFrame 생성 ===")
# 딕셔너리를 DataFrame으로 변환하고 환자ID를 인덱스로 설정
df = pd.DataFrame(patient_data).set_index('환자ID')

# DataFrame 기본 정보 출력
print(f"환자 수: {len(df)}명")
print(f"컬럼 수: {len(df.columns)}개")
print(f"컬럼명: {list(df.columns)}")
print("\n생성된 DataFrame:")
print(df)

# 2. 특정 환자 정보 조회 (loc, iloc 활용)
print("\n=== 2. 특정 환자 조회 ===")
# P003 환자의 모든 정보 조회
print("P003 환자 정보:")
print(df.loc['P003'])

# 첫 번째와 세 번째 환자 정보 조회
print("\n첫 번째, 세 번째 환자 정보:")
print(df.iloc[[0, 2]])

# 모든 환자의 나이와 혈압 정보만 선택
print("\n모든 환자의 나이와 혈압:")
print(df[['나이', '혈압']])

# 3. 조건부 데이터 필터링
print("\n=== 3. 조건부 필터링 ===")
# 고혈압 환자 찾기 (혈압 >= 140)
high_bp_patients = df[df['혈압'] >= 140]
print(f"고혈압 환자 ({len(high_bp_patients)}명):")
print(high_bp_patients[['이름', '혈압']])

# 고령 환자 찾기 (나이 >= 65)
elderly_patients = df[df['나이'] >= 65]
print(f"\n고령 환자 ({len(elderly_patients)}명):")
print(elderly_patients[['이름', '나이']])

# 내과 환자 중 고혈당 환자 찾기 (진료과 == '내과' AND 혈당 >= 120)
internal_high_sugar = df[(df['진료과'] == '내과') & (df['혈당'] >= 120)]
print(f"\n내과 고혈당 환자 ({len(internal_high_sugar)}명):")
print(internal_high_sugar[['이름', '진료과', '혈당']])

# 4. 기본 통계 분석 (value_counts 중심)
print("\n=== 4. 기본 통계 ===")
# 진료과별 환자 수 계산
print("진료과별 환자 수:")
dept_counts = df['진료과'].value_counts()
print(dept_counts)

# 성별 분포 확인
print("\n성별 분포:")
gender_counts = df['성별'].value_counts()
print(gender_counts)

# 5. 보너스: 위험도 등급 추가
print("\n=== 5. 위험도 등급 추가 ===")
# 고혈압이면 '고위험', 아니면 '정상'으로 새 컬럼 생성
df.loc[df['혈압'] >= 140, '위험도'] = '고위험'
df.loc[df['혈압'] < 140, '위험도'] = '정상'

# 위험도별 환자 분포 확인
print("위험도별 환자 분포:")
risk_counts = df['위험도'].value_counts()
print(risk_counts)

# 최종 DataFrame 확인 (새 컬럼 포함)
print("\n최종 DataFrame (위험도 컬럼 추가):")
print(df)
````
