# 실습과제: 당뇨병 데이터 탐색적 분석 연습

병원에서 수집한 당뇨병 예측 데이터를 활용하여 EDA(탐색적 데이터 분석) 기법을 연습해보세요.   
groupby를 활용한 그룹별 분석부터 빈도분석, 교차분석까지 의료 데이터 분석에서 필수적인 기법들을 실습합니다.

## 요구사항

1. **그룹별 분석**
   - 당뇨병 여부별 BMI, 나이의 평균, 표준편차 계산
   - 당뇨병 여부별 BMI 평균 차이 계산

2. **범주형 변수 빈도분석**
   - 성별(Sex) 분포 확인 (빈도와 비율)
   - 교육 수준(Education) 분포 분석  

3. **교차분석 (교차표)**
   - 고혈압과 당뇨병의 연관성 파악
   - 비율 기반 교차표로 패턴 확인

## 힌트

- 그룹별 분석: `df.groupby('그룹컬럼')['타겟컬럼'].agg(['mean', 'std'])`
- 빈도분석: `.value_counts()`, `.value_counts(normalize=True)` 활용
- 교차표: `pd.crosstab(df['컬럼1'], df['컬럼2'])` 사용
- 비율 교차표: `pd.crosstab(df['컬럼1'], df['컬럼2'], normalize='index')`

````python
# 2. 데이터 다운로드
import pandas as pd
import kagglehub

# 당뇨병 건강지표 데이터셋 다운로드 (자동으로 최신 버전 받아옴)
path = kagglehub.dataset_download("alexteboul/diabetes-health-indicators-dataset")

print("데이터가 다운로드된 경로:", path)

# 3. 다운로드된 경로에서 파일 불러오기
df = pd.read_csv(f"{path}/diabetes_binary_health_indicators_BRFSS2015.csv")

df.head()
````

## 💻 구현해보세요



````python
import pandas as pd

# 당뇨병 데이터는 df 변수로 로드됨
# 주요 컬럼: Diabetes_binary, BMI, Sex, Education, HighBP, Age 등

# 1. 그룹별 분석
print("=== 1. 그룹별 분석 ===")
# 당뇨병 여부별 BMI, 나이의 평균과 표준편차
print("당뇨병 여부별 BMI, 나이 통계:")
diabetes_bmi_age_stats=(
    df
    .groupby('Diabetes_binary')[['BMI', 'Age']]
    .agg(['mean', 'std'])
)
diabetes_bmi_age_stats.index=['정상','당뇨병']
print(diabetes_bmi_age_stats.round(2))
# 당뇨병 여부별 BMI 평균 차이 계산
bmi_mean_diff = diabetes_bmi_age_stats['BMI']['mean'].diff().iloc[-1]
print(round(bmi_mean_diff, 2))
print(f"\n📊 그룹 간 차이 분석:")
print(f"평균 BMI 차이: {bmi_mean_diff:.2f} (당뇨병 환자가 더 높음)")
print(f"💡 해석: 당뇨병 환자의 평균 BMI가 {bmi_mean_diff:.1f} 높아 비만과 당뇨병의 연관성 시사")

# 2. 범주형 변수 빈도분석
print("\n=== 2. 빈도분석 ===")
# 성별 분포 (빈도와 비율)
#정답을 봤는데 for문 돌리는것보단 이게 더 간편하고 pandas다운 형식같아서.......
#계산용 column과 출력용 column 분리(원칙)
'''
normalize를 쓸경우
count = df['Sex'].value_counts()
ratio = df['Sex'].value_counts(normalize=True)
'''
gender_stats = (
    df['Sex']
    .value_counts()
    .rename(index={0:"여성",1:"남성"})
    .to_frame('count')
    .assign(
        ratio=lambda x: x['count'] / x['count'].sum())
)
#출력형식을 두가지 방법으로 변환해봤습니다.(1)
gender_stats['count_fmt']=gender_stats['count'].apply(lambda x: f'{x:,}명')
gender_stats['ratio_fmt']=gender_stats['ratio'].apply(lambda x: f'{x:.1%}')
print("성별 분포:")
print(gender_stats[['count_fmt','ratio_fmt']])

print("\n교육 수준 분포:")
edu_stats = (
    df['Education']
    .value_counts()
    .sort_index()
    .to_frame('count')
    .rename(index=lambda x: f'수준 {x}')
    .assign(
        ratio=lambda x: x['count'] / x['count'].sum(),
        #출력형싱 변환 두번째방법, lambda함수 사용
        count_fmt=lambda x: x['count'].map(lambda v: f"{v:,}명"),
        ratio_fmt=lambda x: x['ratio'].map(lambda v: f"{v:.1%}")
    )
)

print(edu_stats[['count_fmt','ratio_fmt']])

# 3. 교차분석 (교차표)
# 고혈압과 당뇨병의 연관성 파악
print("\n=== 3. 교차분석 ===")
print("고혈압 × 당뇨병 교차표 (빈도):")
htn_dm_crosstab = (pd.crosstab(df['HighBP'],df['Diabetes_binary'], margins=True))
htn_dm_crosstab.index = ['정상','고혈압','전체']
htn_dm_crosstab.columns = ['정상','당뇨병','전체']
print(htn_dm_crosstab)
# 비율 기반 교차표로 패턴 확인
print("고혈압 × 당뇨병 교차표 (비율):")
htn_dm_ratio_crosstab = pd.crosstab(df['HighBP'],df['Diabetes_binary'],normalize='index')
htn_dm_ratio_crosstab.index = ['정상','고혈압]']
htn_dm_ratio_crosstab.columns = ['정상','당뇨병']
#display 사용해봄
display(htn_dm_ratio_crosstab.style.format('{:.1%}'))
````

````python
# @title 정답

import pandas as pd

# 1. 그룹별 분석
print("=== 1. 그룹별 분석 ===")

# 당뇨병 여부별 BMI와 나이의 평균, 표준편차
print("당뇨병 여부별 BMI, 나이 통계:")
diabetes_stats = df.groupby('Diabetes_binary')[['BMI', 'Age']].agg(['mean', 'std'])
diabetes_stats.index = ['정상', '당뇨병']
print(diabetes_stats.round(2))

# 당뇨병 여부별 BMI 평균 차이 계산
bmi_by_diabetes = df.groupby('Diabetes_binary')['BMI'].mean()
mean_diff = bmi_by_diabetes[1] - bmi_by_diabetes[0]  # 당뇨병 환자 - 정상인

print(f"\n📊 그룹 간 차이 분석:")
print(f"평균 BMI 차이: {mean_diff:.2f} (당뇨병 환자가 더 높음)")
print(f"💡 해석: 당뇨병 환자의 평균 BMI가 {mean_diff:.1f} 높아 비만과 당뇨병의 연관성 시사")

# 2. 범주형 변수 빈도분석
print("\n=== 2. 빈도분석 ===")

# 성별 분포 (빈도와 비율)
print("성별 분포:")
sex_counts = df['Sex'].value_counts()
sex_ratio = df['Sex'].value_counts(normalize=True)
for category in sex_counts.index:
    count = sex_counts[category]
    ratio = sex_ratio[category]
    gender = "여성" if category == 1 else "남성"
    print(f"  {gender}: {count:,}명 ({ratio:.1%})")

# 교육 수준 분포
print(f"\n교육 수준 분포:")
edu_counts = df['Education'].value_counts().sort_index()
edu_total = len(df)
for level in edu_counts.index:
    count = edu_counts[level]
    ratio = count / edu_total
    print(f"  수준 {level}: {count:,}명 ({ratio:.1%})")


# 3. 교차분석 (교차표)
print("\n=== 3. 교차분석 ===")

# 고혈압과 당뇨병의 교차표
print(f"\n고혈압 × 당뇨병 교차표 (빈도):")
bp_diabetes_crosstab = pd.crosstab(df['HighBP'], df['Diabetes_binary'], margins=True)
bp_diabetes_crosstab.index = ['정상혈압', '고혈압', '전체']
bp_diabetes_crosstab.columns = ['정상', '당뇨병', '전체']
print(bp_diabetes_crosstab)

# 고혈압과 당뇨병의 비율 교차표
print(f"\n고혈압 × 당뇨병 교차표 (비율):")
bp_diabetes_ratio = pd.crosstab(df['HighBP'], df['Diabetes_binary'], normalize='index')
bp_diabetes_ratio.index = ['정상혈압', '고혈압']
bp_diabetes_ratio.columns = ['정상', '당뇨병']
print((bp_diabetes_ratio * 100).round(1))   # 퍼센트로 표시
````
