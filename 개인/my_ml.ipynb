```python
import sqlite3

db_path=r"C:\Users\sdh08\PycharmProjects\PythonProject1\개인\mydb.db"

conn=sqlite3.connect(db_path)

cur=conn.cursor()

cur.execute("""
     create table if not exists users(
        id integer primary key autoincrement
       ,name text
       , age integer
      , email text
     )

""")

sample_data=[('test',35,'user1@aaa.com' )
        ,('hong',20, 'user2@aaa.com')
        ,('kim',10,'user3@aaa.com')
        ,('park',15, 'user4@aaa.com')
        ,('lee',25,'user5@aaa.com')
        ,('kang',30,'user6@aaa.com')
        ]

cur.executemany(" insert into users (name, age, email) values(?,?,?)", sample_data)

conn.commit()

conn.close()
```

```python
pima_columns = ['pregnancies', 'glucose', 'blood_pressure', 'skin_thickness',
                'insulin', 'bmi', 'diabetes_pedigree_function', 'age', 'outcome']
pima_data_url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

# names 파라미터를 사용하여 컬럼명 지정
df = pd.read_csv(pima_data_url, names=pima_columns)
```
