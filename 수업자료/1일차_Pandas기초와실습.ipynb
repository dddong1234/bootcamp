## Pandas란 무엇인가?

Pandas는 Python의 데이터 분석 및 조작을 위한 핵심 라이브러리입니다.  
엑셀의 스프레드시트와 테이블 형식의 데이터 (DataFrame)을 쉽고 직관적으로 다룰 수 있게 해주는 도구입니다.

데이터 분석에서 Pandas가 필수인 이유
- **다양한 데이터 형태 지원**: 숫자, 문자, 날짜 등 혼합된 데이터 처리 가능
- **직관적인 표 형태**: 엑셀과 유사한 방식으로 데이터 조작
- **강력한 데이터 처리**: 대용량 데이터도 빠르고 효율적으로 처리
- **유연한 데이터 변환**: 필터링, 정렬, 그룹화 등 다양한 분석 작업 지원

의료 분야에서는 환자 데이터, 검사 결과, 진료 기록 등의 복합적인 의료 정보를 분석하고 관리하는 데 사용됩니다.

### NumPy vs Pandas

NumPy는 숫자 계산용(수학·배열 연산), Pandas는 표 형태 데이터 분석용(엑셀처럼 행·열 처리)입니다.

**pandas는 numpy의 기반 위에서 만들어진 고급 데이터 분석 도구**로, numpy만으로 분석하기 힘든 데이터 연산을 해줍니다.

## 데이터 구조 생성하기

Pandas의 두 가지 핵심 데이터 구조인 Series와 DataFrame을 이해하고 다양한 방법으로 생성해봅니다.


### Series

Series는 1차원 배열로, 인덱스(라벨)와 값의 쌍으로 구성됩니다.

**Series 특징**
- **구조**: 인덱스(라벨) + 값의 1차원 데이터
- **비유**: 엑셀에서 한 개 열을 선택한 것과 동일
- **활용**: 각 데이터에 의미있는 이름을 붙여 관리 가능

```python
# Series 생성 문법
# pd.Series(데이터, index=라벨목록)
```


````python
# Series 예시: 환자별 수축기 혈압 정보
patient_bp = pd.Series([120, 130, 140, 110, 125],  # 혈압 수치 (값)
                       index=['김철수', '이영희', '박민수', '최지은', '정도영'])  # 환자 이름 (인덱스)

print("환자별 수축기 혈압 (Series)")
patient_bp
````

````python
print(f"Series 타입: {type(patient_bp)}")
print(f"김철수의 혈압: {patient_bp['김철수']} mmHg")  # 인덱스로 특정 값 접근
````

### DataFrame

DataFrame은 2차원 표 형태로, 행과 열로 구성된 구조화된 데이터를 저장합니다.

**DataFrame 특징**
- **구조**: 행(index)과 열(columns)로 구성된 2차원 테이블
- **비유**: 엑셀 스프레드시트와 완전히 동일한 개념
- **활용**: 여러 종류의 정보를 하나의 표로 관리
- **관계**: 여러 개의 Series를 모아놓은 구조

행(row)과 컬럼(column) 은 다음과 같은 의미를 가집니다.

- 행(row): ```하나의 개체(관측 대상, record)``` 를 의미합니다.  
예) 환자 1명, 진료 기록 1건

- 컬럼(column): ```데이터의 속성(특징, feature)``` 을 의미합니다.  
예) 나이, 성별, 혈압, 혈당, BMI 등

#### DataFrame 생성 2가지 방법

DataFrame을 생성하는 주요 방법을 알아봅시다.


- **딕셔너리 방식**: 컬럼 중심, 직관적이고 읽기 쉬움
- **2차원 배열 방식**: 행별 레코드, 코드가 간결하고 속도가 딕셔너리 방식보다  빠름



````python
# DataFrame 생성용 공통 변수
patient_names = ['김철수', '이영희', '박민수', '최지은']
columns = ['나이', '혈압', '혈당']
````

##### 1. 딕셔너리로 생성 (가장 일반적)


````python
# 1. 딕셔너리로 생성 - 컬럼 중심 데이터
patient_dict = {
    '나이': [45, 67, 32, 29],      # 열 단위로 데이터 구성
    '혈압': [120, 140, 110, 115],
    '혈당': [95, 180, 88, 92]
}
df1 = pd.DataFrame(patient_dict, index=patient_names)

print("딕셔너리로 생성한 DataFrame:")
df1
````

##### 2. 2차원 배열로 생성


````python
# 2. 딕셔너리 리스트로 생성 - 행별 딕셔너리
patient_records = [
    [45, 120, 95],    # 김철수 (행)
    [67, 140, 180],   # 이영희
    [32, 110, 88],    # 박민수
    [29, 115, 92]     # 최지은
]
df2 = pd.DataFrame(patient_records, index=patient_names, columns=columns)

print("딕셔너리 리스트로 생성한 DataFrame: ")

df2
````

````python
# DataFrame 예시: 10명 환자의 종합 건강 정보
import pandas as pd

patient_data = {
    '환자ID': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006', 'P007', 'P008', 'P009', 'P010'],  # 환자 식별번호
    '이름': ['김철수', '이영희', '박민수', '최지은', '정도영', '장미래', '오성진', '한소희', '윤태완', '송지원'],  # 환자 성명
    '나이': [45, 67, 32, 29, 12, 41, 35, 102, 28, 63],  # 만 나이
    '성별': ['남', '여', '남', '여', '남', '여', '남', '여', '남', '여'],  # 성별 구분
    '키': [175, 162, 180, 158, 170, 165, 172, 160, 178, 155],  # 키 (cm)
    '체중': [70, 58, 85, 52, 75, 60, 68, 65, 82, 48],  # 체중 (kg)
    '수축기혈압': [120, 140, 110, 115, 135, 125, 130, 145, 118, 120],  # 수축기 혈압 (mmHg)
    '이완기혈압': [80, 90, 70, 75, 85, 82, 88, 95, 78, 92],  # 이완기 혈압 (mmHg)
    '혈당': [95, 180, 88, 92, 110, 102, 98, 156, 89, 134],  # 공복혈당 (mg/dL)
    '흡연': ['비흡연', '금연', '비흡연', '비흡연', '흡연', '비흡연', '흡연', '금연', '비흡연', '흡연'],  # 흡연 상태
    '운동': ['규칙적', '가끔', '안함', '규칙적', '가끔', '규칙적', '안함', '가끔', '규칙적', '안함']  # 운동 습관
}

health_df = pd.DataFrame(patient_data)  # 딕셔너리를 DataFrame으로 변환

print("환자 건강 정보 (DataFrame)")
health_df
````

````python
print(f"타입: {type(health_df)}")
print(f"데이터 크기: {health_df.shape})")  # (행 수, 열 수) 튜플 반환
````

### DataFrame vs Series

DataFrame에서 한 개 컬럼을 선택하면 Series가 되고, 여러 컬럼을 선택하면 DataFrame이 됩니다.

**핵심 문법**
- `df['컬럼명']` → Series (1차원)
- `df[['컬럼명1', '컬럼명2']]` → DataFrame (2차원)  
- DataFrame = 여러 개의 Series로 구성된 구조


````python
print("DataFrame의 한 컬럼은 Series")
age_series = health_df['나이']  # 한 개 컬럼 선택 → Series 반환

print(f"나이 컬럼 타입: {type(age_series)}")
print()

age_series
````

````python
print("DataFrame의 여러 컬럼은 DataFrame")
basic_info_df = health_df[['이름', '나이', '성별']]  # 여러 컬럼 선택 → DataFrame 반환

print(f"기본정보 타입: {type(basic_info_df)}")
print()

basic_info_df

````

## CSV 파일 읽기/쓰기

CSV(Comma-Separated Values)는 데이터 분석의 핵심 파일 형식입니다.

**실무에서 데이터 분석 시 압도적으로 많이 사용되는 방식**으로, 대부분의 데이터 사이언티스들이 일상적으로 활용하는 표준 형식입니다.

**CSV가 데이터 분석에서 널리 사용되는 이유**
- **범용성**: Excel, Google Sheets, 데이터베이스 등 어디서나 읽고 쓸 수 있음
- **호환성**: 운영체제나 소프트웨어에 관계없이 동일하게 동작
- **경량성**: 텍스트 기반으로 용량이 작고 전송 속도가 빠름
- **가독성**: 사람이 직접 읽고 편집할 수 있는 구조
- **표준화**: 웹 크롤링, API 응답, 데이터베이스 내보내기의 기본 형식


**데이터 분석 워크플로우**
```
원시 데이터 수집 → CSV로 저장 → Pandas로 분석 → 결과를 다시 CSV로 내보내기
```


### CSV 파일로 저장하기

## 데이터 탐색 및 구조 속성 파악

데이터 분석의 시작 단계로, 데이터의 전체적인 구조와 특성을 파악하는 탐색적 데이터 분석(EDA) 과정입니다.


### 기본 데이터 탐색

데이터를 처음 받았을 때 가장 먼저 확인해야 할 기본 정보들을 추출하는 방법입니다.

**주요 메서드**
- `len(df)`: 데이터의 행의 갯수 확인
- `head(n)`: 데이터의 처음 n행을 확인하여 데이터 구조 파악
- `tail(n)`: 데이터의 마지막 n행을 확인하여 데이터 끝부분 상태 점검  
- `info()`: 데이터 타입, 결측치, 메모리 사용량 등 전반적 정보 확인
- `describe()`: 숫자형 데이터의 기본 통계량(평균, 표준편차, 최솟값, 최댓값 등) 확인

```python
# 데이터 탐색 문법
# len(df)         # 행 갯수
# df.head(n=5)    # 처음 n행
# df.tail(n=5)    # 마지막 n행  
# df.info()       # 전반적 정보
# df.describe()   # 기본 통계량
```


````python
print(f"총 {len(df)}명의 환자 데이터")
````

````python
# 처음과 끝 데이터 확인
print("처음 5명:")
df.head()  # 기본값은 5행
````

````python
print("마지막 3명:")
df.tail(3)  # 괄호 안 숫자로 행 수 지정
````

### 데이터 구조 속성 파악

데이터를 다루기 전에 데이터의 구조를 파악하는 것이 중요합니다.

**주요 속성**
- `shape`: 배열의 형태 (행, 열 개수) - 환자 수와 특성 개수 확인
- `ndim`: 배열의 차원 수 - 벡터(1차원)인지 행렬(2차원)인지 구분  
- `dtype`: 데이터 타입 - 정수형, 실수형, 문자형 등 확인
- `size`: 전체 요소 개수 - 총 데이터 포인트 수 파악


````python
# 기본 데이터 정보
print(f"전체 환자 수: {df.shape[0]}명")  # 행 개수 = 환자 수
print(f"수집 항목 수: {df.shape[1]}개")   # 열 개수 = 측정 항목 수
print(f"전체 데이터 포인트: {df.size}개")  # 전체 셀 개수
````

````python

# 데이터 타입 정보
data_tyes = df.dtypes  # 각 컬럼의 데이터 타입 확인
data_tyes
````

````python
# 상세 데이터 구조
df.info()    # 컬럼별 타입, 결측치 개수, 메모리 사용량 등 종합 정보
````

### 기본 통계 정보 확인

숫자형 데이터의 분포와 특성을 파악하고, 범주형 데이터의 빈도를 확인합니다.


#### describe() 메서드
- **count**: 결측값이 아닌 데이터 개수
- **mean**: 평균값
- **std**: 표준편차 (데이터 분포 정도)
- **min**: 최솟값
- **25%, 50%, 75%**: 1분위수, 중앙값, 3분위수
- **max**: 최댓값


````python
# 숫자형 데이터 기본 통계 요약
df.describe()  # count, mean, std, min, 25%, 50%, 75%, max 값 출력
````

#### value_counts() 메서드

## 고급 데이터 인덱싱

DataFrame에서 특정 데이터를 선택하는 방법에는 2가지가 있습니다.

#### iloc vs loc

| 구분 | iloc | loc |
|------|------|-----|
| **기준** | 위치(숫자) | 라벨(문자열) or 인덱스 번호 |
| **인덱스** | 0, 1, 2... | 0, 1, 2... 또는 'P001', 'P002'... |
| **슬라이싱** | 끝 미포함 (`0:3` = 0,1,2) | 끝 포함 (`'A':'C'` = A,B,C) |
| **사용법** | `df.iloc[0]` | `df.loc['환자A']` |
| **장점** | 위치가 명확 | 의미가 직관적 |
| **단점** | 순서 변경시 위험 | 라벨이 있어야 함 |


### iloc: 위치 기반 인덱싱

**iloc**은 **i**nteger **loc**ation의 줄임말로, 숫자 위치(0부터 시작)를 기반으로 데이터를 선택합니다.


| 특징 | 설명 | 예시 |
|------|------|------|
| **기본 원리** | 위치 기반 (0부터 시작하는 숫자) | `df.iloc[0]` = 첫 번째 행 |
| **슬라이싱 규칙** | 끝 인덱스 미포함 | `df.iloc[0:3]` = 0,1,2번째 (3개) |
| **음수 지원** | 뒤에서부터 접근 | `df.iloc[-1]` = 마지막 행 |
| **리스트 선택** | 불연속 위치를 한번에 선택 | `df.iloc[[0,2,4]]` = 0,2,4번째 |
| **적합한 상황** | 데이터 위치를 정확히 알 때 | 처음 n개, 마지막 n개 선택 시 |

**iloc 사용 시 주의사항**
- 순수 숫자 인덱스만 사용 (컬럼명 사용 불가)
- 데이터 순서 변경 시 결과가 달라질 수 있음
- 슬라이싱 시 끝 번호는 포함되지 않음
- 리스트 형태로 여러 위치 선택 시 대괄호 안에 리스트 형태로 입력

```python
# iloc 문법
# df.iloc[행위치, 열위치] -> 단일 데이터
# df.iloc[행범위, 열범위] -> 다중 데이터
```


#### iloc로 행 선택


````python
# 1. 단일 행 선택 → Series 반환
print("첫 번째 환자 정보")
first_patient = df.iloc[0]  # 0번째 행 선택
first_patient.head()
````

````python
# 2. 여러 행 선택 (연속) → DataFrame 반환
print("처음 3명 환자 정보 (연속)")
first_three = df.iloc[0:3]  # 0~2번째 행 선택 (3번째 미포함)
first_three.head()
````

````python
# 3. 여러 행 선택 (불연속) → DataFrame 반환
print("1,3,5번째 환자들 (불연속)")
selected_patients = df.iloc[[1, 3, 5]]  # 1,3,5번째 행 선택
selected_patients.head()
````

````python
# 4. 마지막 행 선택 → Series 반환
print("마지막 환자 정보")
last_patient = df.iloc[-1]  # 마지막 행 선택
last_patient.head()
````

````python
# 5. 마지막 3명 선택 → DataFrame 반환
print("마지막 3명 환자")
last_three = df.iloc[-3:]  # 마지막 3개 행 선택
last_three.head()
````

#### iloc로 열 선택


````python
# 1. 단일 열 선택 (위치 기반) → Series 반환
print("두 번째 컬럼 선택 (iloc)")
second_column = df.iloc[:, 1]  # 모든 행, 1번째 열 선택
second_column.head()
````

### loc: 라벨 기반 인덱싱

**loc**은 라벨(인덱스명, 컬럼명)을 기반으로 데이터를 선택합니다.

**Index는 각 행의 번호나 이름을 의미**합니다.



| 특징 | 설명 | 예시 |
|------|------|------|
| **기본 원리** | 라벨 기반 (인덱스명, 컬럼명 사용) | `df.loc['P001']` = P001 환자 |
| **슬라이싱 규칙** | 끝 라벨 포함 | `df.loc['A':'C']` = A,B,C 모두 포함 |
| **조건부 선택** | Boolean 마스크 활용 | `df.loc[df['나이'] > 50]` |
| **리스트 선택** | 불연속 라벨을 한번에 선택 | `df.loc[['P001','P003']]` |
| **적합한 상황** | 의미있는 이름으로 접근할 때 | 특정 환자, 특정 컬럼 선택 시 |

**loc 사용 시 주의사항**
- 인덱스나 컬럼에 해당 라벨이 존재해야 함
- 슬라이싱 시 끝 라벨도 포함됨 (iloc와 다름)
- 조건부 선택 시 Boolean 마스크 결과가 True인 행만 반환
- 숫자 인덱스와 문자 라벨 모두 사용 가능

```python
# loc 문법
# df.iloc[행라벨, 열라벨] -> 단일 데이터
# df.iloc[행범위, 열범위] -> 다중 데이터
# df.loc[조건, 열라벨] -> 조건부 필터링
```

### 조건부 필터링

데이터 분석에서 특정 조건을 만족하는 데이터만 골라내는 것은 매우 중요한 기능입니다.

pandas에서는 조건문을 사용해 원하는 데이터를 쉽게 필터링할 수 있습니다.

**Boolean 마스크 작동 원리**
- **Boolean 마스크**: 조건 결과로 True/False 배열 생성
  
  `df['나이'] >= 50` → `[False, True, True, False, True]`
- **마스크 적용**: True인 위치의 데이터만 추출
  
  `df[df['나이'] >= 50]` → 조건 만족하는 모든 환자 정보

````python
# 단일 조건 - Boolean 마스크 단계별 확인
hypertension_mask = df['수축기혈압'] >= 140

print(f"고혈압 마스크: {hypertension_mask.values}")
print("=" * 30)

# 마스크 적용하여 고혈압 환자 선택
hypertension_patients = df[hypertension_mask]
print(f"고혈압 환자 수: {len(hypertension_patients)}명")

hypertension_patients
````

#### 복합 조건 (여러 조건 조합)

실제 데이터 분석에서는 여러 조건을 동시에 만족하는 데이터를 찾아야 하는 경우가 많습니다.

**복합 조건 연산자**
- `&` (AND): 모든 조건이 True일 때만 True
- `|` (OR): 하나 이상의 조건이 True이면 True
- `~` (NOT): 조건 결과를 반대로 변환

**주의사항**
- 각 조건을 괄호 `()`로 묶어야 함
- 괄호 없이 쓰면 pandas가 의도와 다르게 해석함
- `and`, `or` 대신 `&`, `|` 사용


````python
# 복합 조건 - 여러 조건 조합
age_mask = df['나이'] >= 50
bp_mask = df['수축기혈압'] >= 140
````

````python
# AND 조건: 두 조건을 모두 만족하는 환자

senior_hypertension_patients = df[age_mask &  bp_mask]

print(f"50세 이상 + 고혈압 환자수: {len(senior_hypertension_patients)}명")

senior_hypertension_patients
````

````python
# OR 조건: 하나라도 만족하는 환자
high_risk_patients = df[age_mask | bp_mask]

print(f"50세 이상 또는 고혈압 환자수: {len(high_risk_patients)}명")

high_risk_patients
````

````python
# NOT 조건: 조건을 만족하지 않는 환자
young_normal_patients = df[~age_mask & ~bp_mask]

print(f"50세 미만 + 정상혈압 환자수: {len(young_normal_patients)}명")

young_normal_patients
````

````python
# 3. 조건부로 파생컬럼 추가
df_copy = df.copy() # 원본 데이터프레임 복사

# 혈압상태 컬럼 추가하여 데이터 적재
df_copy.loc[df_copy['수축기혈압'] >= 140, '혈압상태'] = '고혈압'
df_copy.loc[df_copy['수축기혈압'] < 140, '혈압상태'] = '정상'

df_copy.head()
````

````python
# 혈압 상태 분포 확인
bp_status_counts = df_copy['혈압상태'].value_counts()
print("혈압 상태 분포:")
bp_status_counts

````

# DataFrame 데이터 조작

##### 계산 결과 컬럼 추가

**BMI 컬럼 추가**

환자의 체중과 키 정보를 활용하여 BMI 값을 계산한 새로운 컬럼을 추가합니다.

BMI(Body Mass Index, 체질량지수)는 신장과 체중을 이용하여 비만도를 나타내는 지수로, 건강 상태를 평가하는 중요한 지표입니다.

**BMI 계산 공식**
```
BMI = 체중(kg) ÷ (키(m))²
```


````python
# 계산으로 새 컬럼 추가 - BMI 계산
df['BMI'] = df['체중'] / ((df['키'] / 100) ** 2) # 계산 결과: Series타입

df['BMI'] = df['BMI'].round(1) # 소수점 첫째자리까지 표시하여 반올림

df.head(3)

````

##### 조건부 컬럼 생성
**BMI 범주 분류 컬럼 추가**

계산된 BMI 값을 의미있는 범주로 분류하는 새로운 컬럼을 추가합니다.

**BMI 범주 분류 기준**
- 저체중: 18.5 미만
- 정상체중: 18.5~22.9
- 과체중: 23.0~24.9
- 경도비만: 25.0~29.9
- 고도비만: 30.0 이상

````python
# BMI 기준으로 비만도 분류 (조건부 할당 방식)
# 조건에 부합하지 않는 경우 NaN(데이터 없음, 결측치) 할당

df.loc[df['BMI'] < 18.5, 'BMI_범주'] = '저체중'
df.loc[(df['BMI'] >= 18.5) & (df['BMI'] < 23), 'BMI_범주'] = '정상체중'
# df.loc[(df['BMI'] >= 23) & (df['BMI'] < 25), 'BMI_범주'] = '과체중'
df.loc[(df['BMI'] >= 25) & (df['BMI'] < 30), 'BMI_범주'] = '경도비만'
df.loc[df['BMI'] >= 30, 'BMI_범주'] = '고도비만'

print(df['BMI_범주'].value_counts())

print("=" * 30)

print("추가된 DataFrame")
df.head()

````

````python
# 조건 결과 계산하여 추가
df['고위험군'] = df['수축기혈압'] >= 140 # bool값 추가

df.head(3)
````

**조건부 수정 (단일 컬럼)**

**조건부 수정 (다중 컬럼)**

특정 조건을 기준으로 여러 컬럼의 값을 한 번에 수정하는 방식입니다.

```['고위험군', '병원'] 컬럼 값을 조건에 따라 수정```

````python
# 환자 조건에 따라 분석 대상 환자 분류
# - 혈당 126 이상 또는
# - 수축기혈압 140 이상

analysis_target_mask = (df['혈당'] >= 126) | (df['수축기혈압'] >= 140)

# '분석대상', '데이터출처' 컬럼값을 조건에 따라 수정
df.loc[analysis_target_mask, ['분석대상', '데이터출처']] = [True, '내원환자']
df.loc[~analysis_target_mask, ['분석대상', '데이터출처']] = [False, '외래환자']

df[['이름', '혈당', '수축기혈압', '분석대상', '데이터출처']].head(10)
````

##### 여러 행 추가

여러 명의 환자 정보를 동시에 추가할 때는 `concat()` 함수를 사용합니다.

**concat() 함수**  
여러 DataFrame을 연결(concatenate)하는 pandas의 핵심 함수입니다.
- `axis=0` (기본값): 행 방향 연결
- `axis=1`: 열 방향 연결

**ignore_index 옵션**
`ignore_index` 옵션은 DataFrame을 연결할 때 인덱스 처리 방식을 결정하는 파라미터입니다.   
데이터를 추가할 때 인덱스 중복 문제를 해결하고 깔끔한 연속 인덱스를 생성하기 위해 사용됩니다.

- `ignore_index=False` (기본값): 추가되는 데이터의 기존 인덱스를 유지하여 인덱스 중복 발생 가능
- `ignore_index=True` (권장): 추가되는 데이터의 기존 인덱스를 무시하고 새로운 연속 인덱스(0,1,2...) 생성

**권장 이유**
- 인덱스 중복으로 인한 데이터 접근 오류 방지
- 일관된 연속 인덱스로 데이터 관리 용이성 향상
- 추후 데이터 분석 시 예측 가능한 인덱스 구조 제공

## 3. 데이터 삭제

분석에 필요하지 않은 행이나 컬럼을 제거해 **분석 대상만 남기기 위한 데이터 정제 작업**입니다.

### drop() 메서드

**drop() 메서드**
DataFrame에서 불필요한 **행 또는 컬럼을 제거**할 때 사용하는 메서드입니다.  
전처리 과정에서 데이터 구조를 단순화하거나 쓰레기값을 제거할 때 활용합니다.

### 주요 옵션

**inplace 옵션**
원본 DataFrame을 직접 수정할지 여부를 결정합니다.

* `inplace=False` (기본값): 새로운 DataFrame 반환, 원본 유지
* `inplace=True`: 원본 DataFrame 직접 수정하고 `None` 반환

**axis 옵션**
데이터 삭제 방향을 지정하는 옵션으로, DataFrame의 행과 열 중 어느 방향으로 삭제할지를 결정합니다.

* `axis=0` (기본값): 행(인덱스) 삭제
* `axis=1`: 열(컬럼) 삭제


````python
# 1. 원본 DataFrame 보호 방식 (inplace=False, 기본값)
df_without_hospital = df.drop('병원', axis=1)  # 새로운 DataFrame 반환

print(f"원본 컬럼 수: {len(df.columns)}")
print(f"복사본 컬럼 수: {len(df_without_hospital.columns)}")
````

### inplace (원본 보호 vs 원본 직접 수정)


````python
# 2. 원본 DataFrame 직접 수정 방식 (inplace=True)
df_copy = df.copy()  # 복사본 생성
print(f"삭제 전 컬럼 수: {len(df_copy.columns)}")

df_copy.drop('병원', axis=1, inplace=True)  # 원본 직접 수정
print(f"삭제 후 컬럼 수: {len(df_copy.columns)}")
````
