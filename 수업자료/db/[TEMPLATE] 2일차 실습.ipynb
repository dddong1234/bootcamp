- patients
  |컬럼|타입|제약|설명|
  |:--:|:--:|:--:|:--:|
  |patient_id|INT|PK, AUTO_INCREMENT|환자 식별 기본키|
  |name|VARCHAR(50)|NOT NULL|환자 이름|
  |gender|ENUM('M', 'F')||성별 (남/여)|
  |birth_date|DATE||생년월일|

- diagnoses
  |컬럼|타입|제약|설명|
  |:--:|:--:|:--:|:--:|
  |diag_id|INT|PK, AUTO_INCREMENT|진단 ID|
  |patient_id|INT|FK -> patients(patient_id)|어떤 환자에게 내린 진단인지 연결|
  |diagnosis_name|VARCHAT(50)||진단명|
  |diag_date|DATE||진단 날짜|

 - visits
    |컬럼|타입|제약|설명|
    |:--:|:--:|:--:|:--:|
    |visit_id|INT|PK, AUTO_INCREMENT|방문 기록 ID|
    |patient_id|INT|FK -> patients(patient_id)|방문한 환자|
    |visit_date|DATETIME||방문한 날짜/시간|
    |doctor|VARCHAR(50)||진료한 의사 이름|

````python
%%sql
%%sql
CREATE TABLE patients(
    patient_id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    gender ENUM('M','F'),
    birth_date DATE
);

CREATE TABLE diagnosis(
    diag_id INT AUTO_INCREMENT PRIMARY KEY,
    patient_id INT,
    diagnosis_name VARCHAR(50),
    diag_date DATE,
    CONSTRAINT fk_diagnosis_patient
        FOREIGN KEY (patient_id) REFERENCES patients(patient_id)
);

CREATE TABLE visits(
    visit_id INT PRIMARY KEY AUTO_INCREMENT,
    patient_id INT,
    visit_date DATETIME,
    doctor VARCHAR(50),
    CONSTRAINT fk_visits_patient
        FOREIGN KEY (patient_id) REFERENCES patients(patient_id)
);
````

> 1. 고혈압 환자 찾기: 진단명이 Hypertension인 환자 찾기


````python
%%sql
%%sql
select name, patients.patient_id, gender, diagnosis_name
from patients
left join diagnosis
    on patients.patient_id = diagnosis.patient_id
where diagnosis_name = 'Hypertension'
````

> 2. 대사 질환 위험군 찾기:
> - Diabetes 당뇨
> - Hypertension 고혈압

````python
%%sql
%%sql
select name, patients.patient_id, gender, diagnosis_name
from patients
left join diagnosis
    on patients.patient_id = diagnosis.patient_id
where diagnosis_name = 'Hypertension' or diagnosis_name = 'Diabetes'
````

> 3. 최근에 방문한 고위험 환자 TOP 3 찾기:
> - Diabetes 당뇨
> - Hypertension 고혈압
> - 중 하나라도 있으면 고위험 환자로 분류
> - 최근: 방문 날짜 기준

````python
%%sql
%%sql
select name, patients.patient_id, gender, diagnosis_name, visits.visit_date
from patients
join diagnosis
    on patients.patient_id = diagnosis.patient_id
join visits
    on patients.patient_id = visits.patient_id
where diagnosis_name in ('Hypertension','Diabetes')
order by visits.visit_date desc
limit 3
````
