### 라이브러리 import

```python
import pandas as pd
import numpy as npv
```

### Step 1. 데이터 불러오기 (Load Data)

```python
pima_columns = ['pregnancies', 'glucose', 'blood_pressure', 'skin_thickness',
                'insulin', 'bmi', 'diabetes_pedigree_function', 'age', 'outcome']
pima_data_url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
df = pd.read_csv(pima_data_url, names=pima_columns)
```

### Step 3. 데이터 선택 (Select Data)

```python
# 1. 열 선택
bmi_age_df = df[['bmi', 'age']] # DataFrame
display(bmi_age_df)
```

```python
# 2. 행 필터링
# 나이가 50세 이상인 환자 선택
over_50 = df[df['age'] >= 50]
display(over_50)
```

```python
# 3. .loc 사용
display(df.loc[0:2, ['glucose', 'outcome']])
```

```python
# 4. .iloc 사용
display(df.iloc[0:2, [2,3,4]])
```

### Step 4. 데이터 정제 (Clean Data)

```python
# outcome을 문자열로 변환해보기
print(f"\n 타입 변환 전 outcome 타입: {df['outcome'].dtype}")
df['outcome'] = df['outcome'].astype('str')
print(f"\n 타입 변환 후 outcome 타입: {df['outcome'].dtype}")
```

### Step 5. 데이터 재구조화 (Restructure)

```python
# 1. 정렬 (Sorting) - bmi 기준
df_sorted = df.sort_values(by='bmi', ascending=False)
print("\n1. bmi 기준 내림차순 정렬 (Top 3):")
print(df_sorted[['bmi', 'age']].head())
```

```python
# 2. 컬럼 이름 변경 (Renaming)
# pregnancies -> preg_count 로 변경 예시
df_renamed = df.rename(columns={'pregnancies': 'preg_count'})
display(df_renamed)
```

### Step 6. 데이터 결합 (Combine)

```python
# 실습용 분할
df_part1 = df.iloc[:2, :2]
display(df_part1)
```

```python
df_part2 = df.iloc[2:4, :2]
display(df_part2)
```

```python
df_concat = pd.concat([df_part1, df_part2], axis=0)
display(df_concat)
```

### Step 6. 그룹 분석 및 통계 (GroupBy & Statistics)

```python
# 원래 데이터 타입인 int로 복구 후 분석
df['outcome'] = df['outcome'].astype(int)
```

```python
# 1. GroupBy: outcome(당뇨 발병 여부)별 평균
# 수정된 소문자 컬럼명 사용
grouped = df.groupby('outcome').mean()
display(grouped[['glucose', 'bmi', 'age']])
```

```python
# 2. Aggregation - outcome별 glucose 요약 통계
agg_res = df.groupby('outcome')['glucose'].agg(['mean', 'max', 'min'])
display(agg_res)
```

```python
# 3. Apply 함수 적용

# bmi가 30 이상이면 'Obese', 아니면 'Normal'
def classify_bmi(bmi_val):
    if bmi_val >= 30:
        return '비만'
    else:
        return '정상'

df['bmi_class'] = df['bmi'].apply(classify_bmi)
df[['bmi', 'bmi_class']].head()
```

### Step 7. 파생 변수 생성 (Assign)

```python
# insulin 값을 2배로 늘린 파생변수 생성
df_new = df.assign(insulin_double = lambda x: x['insulin'] * 2)
df_new[['insulin', 'insulin_double']].head()
```
