# 실습과제: 환자 혈압 분류하기

병원에서 제공하는 환자 혈압 데이터를 활용하여 혈압 분류, 통계 분석, 고위험군 관리를 수행하는 프로그램을 작성해보세요.

## 요구사항

1. **혈압 분류 기준**
   - 정상: 수축기 < 120
   - 고혈압 전단계: 120 ≤ 수축기 < 140  
   - 1기 고혈압: 140 ≤ 수축기 < 160
   - 2기 고혈압: 수축기 ≥ 160

2. **기능 구현**
   - 각 환자별 혈압 분류 및 위험도 평가
   - 응급 처치가 필요한 환자 즉시 감지
   - 혈압 등급별 통계 및 전체 환자 현황 분석

3. **출력 형식**
```
=== 개별 환자 혈압 분류 ===
김환자(65세): 145 mmHg - 1기 고혈압 (약물 치료 필요)
이환자(45세): 125 mmHg - 고혈압 전단계 (생활습관 개선)
박환자(32세): 115 mmHg - 정상 (정기 검진)
최환자(28세): 135 mmHg - 고혈압 전단계 (생활습관 개선)
정환자(70세): 190 mmHg - 2기 고혈압 (즉시 치료 필요)

=== 응급 처치 대상 ===
🚨 응급: 정환자 (190 mmHg) - 즉시 치료 필요

=== 혈압 등급별 통계 ===
정상: 2명 (40.0%)
고혈압 전단계: 1명 (20.0%)  
1기 고혈압: 1명 (20.0%)
2기 고혈압: 1명 (20.0%)

=== 전체 현황 요약 ===
총 환자 수: 5명
평균 혈압: 142.0 mmHg
고혈압 환자: 2명 (40.0%)
응급 처치 대상: 1명
```

````python
# 기본 데이터 - 환자 혈압 데이터베이스
patients = [
    {"name": "김환자", "age": 65, "systolic": 145},
    {"name": "이환자", "age": 45, "systolic": 125},
    {"name": "박환자", "age": 32, "systolic": 115},
    {"name": "최환자", "age": 28, "systolic": 135},
    {"name": "정환자", "age": 70, "systolic": 190}
]
````

````python
patients = [
    {"name": "김환자", "age": 65, "systolic": 145},
    {"name": "이환자", "age": 45, "systolic": 125},
    {"name": "박환자", "age": 32, "systolic": 115},
    {"name": "최환자", "age": 28, "systolic": 135},
    {"name": "정환자", "age": 70, "systolic": 190}
]

htn_class = [
    "정상",
    "고혈압 전단계",
    "1기 고혈압",
    "2기 고혈압"
]

htn_action = {
    "정상": "정기 검진",
    "고혈압 전단계": "생활습관 개선",
    "1기 고혈압": "약물 치료 필요",
    "2기 고혈압": "즉시 치료 필요"
}

htn_list = {"정상":[],"고혈압 전단계":[],"1기 고혈압":[],"2기 고혈압":[]}

def classify_htn(patient):
    systolic = patient["systolic"]

    if systolic < 120:
        return 0
    elif 120 <= systolic < 140:
        return 1
    elif 140 <= systolic < 160:
        return 2
    else:
        return 3

def classify_emergency(patient):
    systolic = patient["systolic"]
    return systolic >= 180


print("=== 개별 환자 혈압 분류 ===")
htn_count:int = 0
em_count:int = 0
for patient in patients:
    if patient["systolic"] >= 140:
        htn_count += 1
    if patient["systolic"] >= 180:
        em_count += 1
    htn_code = classify_htn(patient)
    htn_list[htn_class[htn_code]].append(patient)
    name = patient["name"]
    age = patient["age"]
    systolic = patient["systolic"]
    print(f"{name}({age}세): {systolic} mmHg - {htn_class[htn_code]} ({htn_action[htn_class[htn_code]]})")

print("\n")
print("=== 응급 처치 대상 ===")
for patient in patients:
    if classify_emergency(patient):
        name = patient["name"]
        systolic = patient["systolic"]
        print(f'🚨 응급 : {name}({systolic} mmHg) - 즉시 치료 필요')
        break

print("\n")

print("=== 혈압 등급별 통계 ===")
for stage, patients_by_stage in htn_list.items():
    print(f"{stage}: {len(patients_by_stage)}명 ({len(patients_by_stage)/len(patients):.1%})")

print("\n")

print("=== 전체 현황 요약 ===")
print(f"총 환자수: {len(patients)}명")
systolic_list = [int(patient["systolic"]) for patient in patients]
print(f"평균 혈압: {sum(systolic_list)/len(systolic_list):.1f} mmHg")
print(f"고혈압 환자: {htn_count}명 ({htn_count/len(patients):.1%})")
print(f"응급 처치 대상: {em_count}명")
````

````python
# @title 정답

# 환자 혈압 데이터베이스
patients = [
    {"name": "김환자", "age": 65, "systolic": 145},
    {"name": "이환자", "age": 45, "systolic": 125},
    {"name": "박환자", "age": 32, "systolic": 115},
    {"name": "최환자", "age": 28, "systolic": 135},
    {"name": "정환자", "age": 70, "systolic": 190}
]

# 1. 개별 환자 혈압 분류
print("=== 개별 환자 혈압 분류 ===")
for patient in patients:
    name = patient["name"]  # 환자 이름
    age = patient["age"]  # 환자 나이
    systolic = patient["systolic"]  # 수축기 혈압 (mmHg)

    # 혈압 분류 및 권고사항
    if systolic < 120:
        category = "정상"
        recommendation = "정기 검진"
    elif systolic < 140:
        category = "고혈압 전단계"
        recommendation = "생활습관 개선"
    elif systolic < 160:
        category = "1기 고혈압"
        recommendation = "약물 치료 필요"
    else:
        category = "2기 고혈압"
        recommendation = "즉시 치료 필요"

    print(f"{name}({age}세): {systolic} mmHg - {category} ({recommendation})")

# 2. 응급 처치 대상 즉시 감지
print("\n=== 응급 처치 대상 ===")
emergency_found = False

for patient in patients:
    name = patient["name"]  # 환자 이름
    systolic = patient["systolic"]  # 수축기 혈압 (mmHg)

    if systolic >= 180:  # 응급 상황 기준
        print(f"🚨 응급: {name} ({systolic} mmHg) - 즉시 치료 필요")
        emergency_found = True
        break  # 응급 환자 발견시 즉시 처리

if not emergency_found:
    print("응급 처치가 필요한 환자가 없습니다.")

# 3. 혈압 등급별 통계 계산
print("\n=== 혈압 등급별 통계 ===")

# 각 등급별 환자 수 계산
normal_count = len([p for p in patients if p["systolic"] < 120])
prehyper_count = len([p for p in patients if 120 <= p["systolic"] < 140])
stage1_count = len([p for p in patients if 140 <= p["systolic"] < 160])
stage2_count = len([p for p in patients if p["systolic"] >= 160])

total_patients = len(patients)

# 비율 계산 및 출력
print(f"정상: {normal_count}명 ({normal_count/total_patients*100:.1f}%)")
print(f"고혈압 전단계: {prehyper_count}명 ({prehyper_count/total_patients*100:.1f}%)")
print(f"1기 고혈압: {stage1_count}명 ({stage1_count/total_patients*100:.1f}%)")
print(f"2기 고혈압: {stage2_count}명 ({stage2_count/total_patients*100:.1f}%)")

# 4. 전체 현황 요약
print("\n=== 전체 현황 요약 ===")

# 혈압 값들 추출
all_systolic = [patient["systolic"] for patient in patients]

# 통계 계산
total_patients = len(patients) # 총 환자 수
avg_systolic = sum(all_systolic) / len(all_systolic) # 평균 혈압
hypertension_patients = len([p for p in patients if p["systolic"] >= 140]) # 고혈압 환자 수
emergency_patients = len([p for p in patients if p["systolic"] >= 180]) # 응급 처치 대상 수

print(f"총 환자 수: {total_patients}명")
print(f"평균 혈압: {avg_systolic:.1f} mmHg")
print(f"고혈압 환자: {hypertension_patients}명 ({hypertension_patients/total_patients*100:.1f}%)")
print(f"응급 처치 대상: {emergency_patients}명")
````
