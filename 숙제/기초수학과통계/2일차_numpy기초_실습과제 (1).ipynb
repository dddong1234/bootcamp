````python
import numpy as np

patients_data = np.array([
    [80, 45.0, 250, 490, 780],   # 환자1: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [75, 38.0, 200, 220, 300],   # 환자2: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [35, 23.0, 115, 90, 180],   # 환자3: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [60, 30.0, 160, 140, 250],   # 환자4: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [40, 25.0, 120, 100, 200]    # 환자5: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
])

risk_weights = np.array([0.05, 0.08, 0.02, 0.03, 0.01])

stages = ["낮음", "보통", "높음", "매우 높음"]
clinical_recommendation = ["정기검진 유지", "생활습관 개선", "전문의 상담", "즉시 치료"]


# 환자들의 데이터에 가중합 계산을 해 반환
def calculate_risk(patients_data,risk_weights):
    risks = np.dot(patients_data, risk_weights)
    return risks

# 반복되는 구문들을 list화 하여 반환
def get_risk_strings(risks):
    str_list = []
    for index, risk in enumerate(risks):
        str_list.append(f"환자{index+1}: {risk:.1f}점")
    return str_list

# 1단계 출력을 위한 함수
def print_risk_strings(str_list):
    for string in str_list:
        print(string)

# 환자들의 위험도를 토대로 단계별 dict에 삽입후 반환, stages list와 clinical_recommendation list를 한번에 제어할 수 있는 index_list 반환
def get_stage_index_list(risks):
    risk_stage = {
        "낮음": [],
        "보통": [],
        "높음": [],
        "매우 높음": []
    }
    index_list = []

    for risk in risks:
        if risk < 15:
            stage = "낮음"
            index = 0
        elif 15 <= risk < 25:
            stage = "보통"
            index = 1
        elif 25 <= risk < 35:
            stage = "높음"
            index = 2
        else:
            stage = "매우 높음"
            index = 3
        risk_stage[stage].append(risk) # 만약 patient 객체가 있다면 risk 대신 patient객체를 넣고싶음
                                       #현재는 쓰이지 않는 dict이지만 미래 확장성을 생각한다면 만들어도 좋을듯
        index_list.append(index)
    return index_list, risk_stage

# 2단계 출력을 위한 함수
def print_patients_summary(index_list,str_list):
        for i, risk_str in enumerate(str_list):
            stage = stages[index_list[i]]
            recommendation = clinical_recommendation[index_list[i]]
            print(f"{risk_str} ({stage}) → {recommendation}")

# 극값 index 반환
def get_max_min_index(risks):
    max_risk_index = np.argmax(risks)
    min_risk_index = np.argmin(risks)
    return max_risk_index, min_risk_index

# 3단계 출력을 위한 함수
def print_max_min_patient(risks):
    max_risk_index, min_risk_index = get_max_min_index(risks)
    print(f"가장 위험한 환자: 환자{max_risk_index+1} ({risks[max_risk_index]:.1f}점)")
    print(f"가장 안전한 환자: 환자{min_risk_index + 1} ({risks[min_risk_index]:.1f}점)")



risks = calculate_risk(patients_data,risk_weights)
index_list, risk_stage = get_stage_index_list(risks)
str_list= get_risk_strings(risks)

print("=== 환자 위험도 통합 평가 시스템 ===\n")
print("=== 1단계: 전체 환자 위험도 계산 ===")
print_risk_strings(str_list)
print("\n=== 2단계: 위험도 등급 분류 ===")
print_patients_summary(index_list,str_list)
print("\n=== 3단계: 최고/최저 위험 환자 ===")
print_max_min_patient(risks)





````

````python
# @title 정답

import numpy as np

# 기본 데이터

# 환자 종합 검진 데이터
patients_data = np.array([
    [80, 45.0, 250, 490, 780],   # 환자1: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [75, 38.0, 200, 220, 300],   # 환자2: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [35, 23.0, 115, 90, 180],   # 환자3: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [60, 30.0, 160, 140, 250],   # 환자4: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
    [40, 25.0, 120, 100, 200]    # 환자5: [나이, BMI, 수축기혈압, 혈당수치, 콜레스테롤]
])

# 각 지표별 위험도 가중치 (훨씬 작게 조정)
risk_weights = np.array([0.05, 0.08, 0.02, 0.03, 0.01])  # [나이, BMI, 혈압, 혈당, 콜레스테롤]

print("=== 환자 위험도 통합 평가 시스템 ===")

# 1단계: 전체 환자 위험도 계산 (행렬곱 사용)
print("\n=== 1단계: 전체 환자 위험도 계산 ===")
# 행렬곱을 사용해 5명 환자의 위험도를 한 번에 계산
all_risks = np.dot(patients_data, risk_weights)

# 각 환자별 위험도 점수 출력
for i, risk_score in enumerate(all_risks, 1):
    print(f"환자{i}: {risk_score:.1f}점")

# 2단계: 위험도 등급 분류
print("\n=== 2단계: 위험도 등급 분류 ===")
for i, risk_score in enumerate(all_risks, 1):
    # 위험도 점수에 따른 등급 분류
    if risk_score < 15:
        level = "낮음"
        recommendation = "정기검진 유지"
    elif risk_score < 25:
        level = "보통"
        recommendation = "생활습관 개선"
    elif risk_score < 35:
        level = "높음"
        recommendation = "전문의 상담"
    else:
        level = "매우 높음"
        recommendation = "즉시 치료"

    print(f"환자{i}: {risk_score:.1f}점 ({level}) → {recommendation}")

# 3단계: 최고/최저 위험 환자 찾기
print("\n=== 3단계: 최고/최저 위험 환자 ===")

# np.argmax(): 배열에서 최대값의 인덱스(위치)를 반환하는 메서드
max_patient_index = np.argmax(all_risks)  # 가장 높은 위험도의 인덱스
max_risk = all_risks[max_patient_index]   # 해당 인덱스의 위험도 값

# np.argmin(): 배열에서 최소값의 인덱스(위치)를 반환하는 메서드
min_patient_index = np.argmin(all_risks)  # 가장 낮은 위험도의 인덱스
min_risk = all_risks[min_patient_index]   # 해당 인덱스의 위험도 값

# 인덱스는 0부터 시작하므로 환자 번호는 +1 해줌
print(f"가장 위험한 환자: 환자{max_patient_index + 1} ({max_risk:.1f}점)")
print(f"가장 안전한 환자: 환자{min_patient_index + 1} ({min_risk:.1f}점)")
````
